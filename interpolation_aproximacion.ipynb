```python
# Import necessary libraries
import matplotlib.pyplot as plt
import math, gmpy2
import numpy as np
from gmpy2 import  mpz, mpfr
import gmpy2 
from algorithm import *
from interpolation_approximation import *

# Set precision
e = 100

# Define coordinate points
coords = [
    (1, 1),
    (2, 4),
    (3, 9)
]
```

```python
# Create instances of interpolation/approximation classes
l1 = LagrangePolynomial(coords)
a1 = Approximation(coords)
ls1 = LinearInterpolationSpline(coords)

# Perform calculations
l1.compute()
a1.compute()

# Print results
print("Lagrange polynomial:", l1.function_str)
print("Approximation:", a1.function_str)

print("\nLinear Interpolation Spline")
ls1.print_result()
```

```python
# Determine the minimum and maximum x values among the coordinates
x_min = min(coord[0] for coord in coords)
x_max = max(coord[0] for coord in coords)

# Sort coordinates in ascending order by x
coords.sort(key = lambda x: x[0])

# Generate 100 points between x_min and x_max
x = np.linspace(x_min, x_max, 100)

# Calculate the y values for each element of the x vector using the methods
fig, ax = plt.subplots()

ax.plot(x, l1.function(x), label = "Lagrange polynomial")
ax.plot([p[0] for p in coords], [p[1] for p in coords], "ro", label = "coordinates")
ax.plot([p[0] for p in coords], [p[1] for p in coords], "-", label = "Linear Interpolation Spline")
ax.plot(x, a1.function(x), label = "approximation")

# Label the plot
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.show()
```
