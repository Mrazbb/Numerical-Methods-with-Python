```python
import matplotlib.pyplot as plt
import math, gmpy2
import numpy as np
from gmpy2 import mpz, mpfr
import gmpy2
from algorithm import *  # Adjust if needed (assuming base class is named Algorithm)
from integral import *   # This imports the translated Integral, TrapezoidalMethod, RectangularMethod

# Example function:
function_example = lambda x: x**3 + 1
a = '2'  # Lower bound
b = '3'  # Upper bound
e = 7    # Number of decimal places
```

```python
# Create instances of both methods and compute:
rect_method = RectangularMethod(function_example, a, b, e)
trap_method = TrapezoidalMethod(function_example, a, b, e)

rect_method.compute()
trap_method.compute()

print("Rectangular method")
rect_method.print_result()

print("\nTrapezoidal method")
trap_method.print_result()
```

```python
# Compare the accuracy progression:
fig, ax = plt.subplots()

ax.plot(np.arange(1, rect_method.accuracy.size + 1),
        rect_method.accuracy, linewidth=2.0,
        label="Rectangular")
ax.plot(np.arange(1, trap_method.accuracy.size + 1),
        trap_method.accuracy, linewidth=2.0,
        label="Trapezoidal")

plt.xlabel('Iterations')
plt.ylabel('Accuracy (log base 10 of difference)')
plt.grid(True)
plt.legend()
plt.show()
```
