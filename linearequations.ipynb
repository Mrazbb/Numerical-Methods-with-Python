```python
import math
import numpy as np
from gmpy2 import is_prime, mpz, mpfr
import gmpy2
import matplotlib.pyplot as plt

# Import the solvers from the local file (previously known as 'vypocet_srl.py').
# Make sure the filename and class names below match the refactored code.
from linearequations import JacobiMethod, GaussSeidelMethod

# Define the matrix A and vector b as lists of strings,
# which will be converted to mpfr in the solver.
A = [
    ['6', '2', '-1'],
    ['1', '3', '1'],
    ['2', '-1','9']
]
b = ['9', '11', '-13']

# Define the number of decimal places used for high-precision arithmetic.
precision_in_decimals = 10
```

```python
# Create instances of each method.
jacobi_solver = JacobiMethod(A, b, precision_in_decimals)
gauss_seidel_solver = GaussSeidelMethod(A, b, precision_in_decimals)

# Perform the computations.
jacobi_solver.compute()
gauss_seidel_solver.compute()

print("Jacobi Method:")
jacobi_solver.print_solution()

print("\nGauss-Seidel Method:")
gauss_seidel_solver.print_solution()
```

```python
# Plot the accuracy (in decimal digits) across iterations for each method.
fig, ax = plt.subplots()

ax.plot(np.arange(1, jacobi_solver.accuracy.size + 1), jacobi_solver.accuracy, linewidth=2.0, label="Jacobi")
ax.plot(np.arange(1, gauss_seidel_solver.accuracy.size + 1), gauss_seidel_solver.accuracy, linewidth=2.0, label="Gauss-Seidel")

plt.xlabel('Iterations')
plt.ylabel('Accuracy (number of decimal places)')
plt.grid(True)
plt.legend()
plt.show()
```
